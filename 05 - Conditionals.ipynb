### **3Ô∏è‚É£ –ö–æ—Ä–æ—Ç–∫–æ–µ –∑–∞–º—ã–∫–∞–Ω–∏–µ (Short-circuit evaluation) –≤ Julia**

- –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ, –≤ –≤—ã—Ä–∞–∂–µ–Ω–∏–∏ `a || b` **–µ—Å–ª–∏ `a == true`**, —Ç–æ –≤–µ—Å—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç —É–∂–µ **–±—É–¥–µ—Ç `true`**, –∏ `b` —Å–Ω–æ–≤–∞ **–Ω–µ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è**.
